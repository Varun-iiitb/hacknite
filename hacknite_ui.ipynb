```python
import gradio as gr
import base64
from gtts import gTTS
import tempfile
```

```python
# Read local image and encode it to base64
image_path = "/home/varun/hacknite/git_1/Pasted_image.png"
with open(image_path, "rb") as img_file:
    encoded_image = base64.b64encode(img_file.read()).decode()

# Google Font + Custom CSS
custom_html = f"""
<link href="https://fonts.googleapis.com/css2?family=Roboto+Mono&display=swap" rel="stylesheet">
<style>
    /* Ensure the background image is applied */
    .gradio-container {{
        background-image: url("data:image/png;base64,{encoded_image}");
        background-size: cover;
        background-repeat: no-repeat;
        background-position: center;
        padding: 40px;
        color: white;
        min-height: 100vh;
        font-family: 'Roboto Mono', monospace;
    }}

    /* Title Style */
    .custom-title {{
        font-size: 36px;
        color: white !important;  /* Force white color */
        text-align: center;
        margin-bottom: 10px;
    }}

    /* Subtitle Style */
    .custom-subtitle {{
        font-size: 24px;
        color: yellow !important;  /* Force yellow color */
        text-align: center;
        margin-bottom: 30px;
    }}

    /* Label Style */
    label {{
        font-size: 16px;
        color: white !important;  /* Force white color */
    }}

    /* Textbox Input Style */
    input, textarea {{
        font-size: 16px;
        color: white !important;  /* Force white color */
        background-color: rgba(0, 0, 0, 0.5) !important;  /* Optional: dark background for input fields */
        border: 1px solid white !important;  /* White border */
    }}

    /* Button Style */
    .gr-button {{
        font-size: 18px;
        font-weight: bold;
        color: white !important;  /* Force white color */
        border: 1px solid white !important;
        background-color: rgba(0, 0, 0, 0.7) !important;  /* Optional: dark background for button */
    }}

    /* Button Hover Style */
    .gr-button:hover {{
        background-color: rgba(255, 255, 255, 0.3) !important;  /* Button hover effect */
    }}
</style>
"""

def chat_with_voice(user_input):
    # Simple chatbot logic
    if "anxious" in user_input.lower():
        response = "I'm here for you. Let's take a few deep breaths together."
    else:
        response = "Thank you for sharing. You are not alone."

    # Convert to speech using gTTS
    tts = gTTS(response)
    temp_file = tempfile.NamedTemporaryFile(suffix=".mp3", delete=False)
    tts.save(temp_file.name)

    return response, temp_file.name

with gr.Blocks() as demo:
    gr.HTML(custom_html)
    gr.HTML('<div class="custom-title">ðŸ§˜ Mental Health Chatbot with Voice</div>')
    gr.HTML('<div class="custom-subtitle">Type how you\'re feeling. The chatbot will reply with empathy â€” and voice.</div>')

    name = gr.Textbox(placeholder="How are you feeling?")
    output = [gr.Textbox(label="Bot says"), gr.Audio(label="Voice Response", type="filepath")]  # Corrected here
    greet_btn = gr.Button("submit")
    greet_btn.click(fn=chat_with_voice, inputs=name, outputs=output)

    # Add examples
    demo.examples = [["I'm feeling really down today... like nothing's going right."],
                     ["I am struggling to cope up with my classes"],
                     ["I am confused on what college and stream to choose"]]

demo.launch()
```
